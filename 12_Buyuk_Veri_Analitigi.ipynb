* Apache Spark küme üzerinde hızlı ve genel amaclı bilgi işleme sistemidir.

## Apache Spark Nedir?

* MapReduce modelinde yer alan disk bazlı çalışma sisteminin yarattıgı maliyetten dolayı ortaya çıkmıştır
* Apache Hadoop ' a göre 100 kat daha hızlı çalışmaktadır
* Java,,Scala,Python ve R ile uygulama geliştirilebilir
* Genelleştiricidir: Spark SQL,Spark MLlib,Spark Streaming,GraphX aynı uygulamada kullanılabilir.

## Apache Spark'ın Bileşenleri

* Spark Core ve RDD's
    - Hadoop Common a benzer olarak spark core; Apache Sparkın temel yapısını olusturan bir kütüphanedir.Hafıza yönetimi, görevlerin dagıtılması, hata kurtarma, saklama ve dosya işlemlerine erişim gibi temel bileşenler bu Spark Core içeirinde  yer alır.
    - Apache Spark'ın programlama modeli  verinini bellek içi (in memory) tutularak paralel işlenmesini ifade eden RDD's dir.Türkçesi dayanıklı dağıtık datasetler demektir.
    
* Spark SQL
    - SQL veya bir dataframe apı yı kullanrak sparkaki yapılandırılmıs veriyi sorgulama imkanı saglar.
* Spark MLlib
    - Apache Spark'ın ölçeklenebilir makine öğrenmesi kütüphanesidir. Apache spark bellek içi çalışma mantıgından dolayı iteratif işlemler  barındıran veri madenciligi ve makine öğrenmesi uygulamaları için uygun bir çözümdür.
* Spark Streaming
    - Akan verinin ölçeklenebilir, yüksek haicmli ve hata toleranslı olarak  işlenmesine imkan saglayan temel bir spark api uzantısıdır.
    - Bir çok farklı kaynaktan gelen veri Spark Streaming e entegre edilebilir ve Map,Reduce,Join,Windows gibi yüksek seviyeli fonksiyonlar ile işlenebilir.
    - İşlenen veriler canlı ekranlara Spark Streaming ile aktarılabilir.
* GraphX
    - Grafikelr ve paralel grafik tabanlı hesaplamalar için kullanılan bir diğer bileşendir. 

* Apache Spark 'ın çalışması  3 aşamadan olusur;
    - RDD
        - Önce bir RDD's olusturulur.
        - Ram üzerinde gerçekleştirir.
        - Diske yazma işlemi gerçekleştirmez
    - Transformation---(Apache Hadoop da Map işlemine denk gelir)
        - RDD's ler üzerinden yeni RDD's ler üretilebilir.
        - Yani veri bellek içinde çeşitli dönüştürme işlemlerinde gidip Harddisck e yazılmaz.Inmemory olarak tutulup üzerinde işlemler gercekleştirilir.
    - Action---(Apache Hadoop da Reduce işlemine denk gelir)
        - Transformasyon işlemi sonrasında üretilen RDD's ler işlenir ve toplanarak sonuçlar elde edilir.
        - Action basamagı tetiklenmeden execution işlemi gerçekleşmemektedir.Yani gidipte hemen harddick e yazmaz.

# GBM ile Müşteri Terk Modellemesi

```python
from pyspark.ml.classification import GBTClassifier
```

```python

# model nesnesi olusturuldu
gbm=GBTClassifier(maxIter=10,featuresCol="features",labelCol="label")
```

```python
# modelin kurulması

gbm_model=gbm.fit(train_df)
```

```python
y_pred=gbm_model.transform(test_df)
```

```python
y_pred
```

```python
ac=y_pred.select("label","prediction")# gercek degerleri(label) ve tahmin edilen degerleri(prediction) sectik
```

```python
ac.filter(ac.label==ac.prediction).count()/ac.count()# accuracy degeri ulasmıs olduk
```

# model tuning işlemlerinin yapılması

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder,CrossValidator
```

```python
evaluator=BinaryClassificationEvaluator()
paramGrid=(ParamGridBuilder()
          .addGrid(gbm.maxDepth,[2,4,6])
          .addGrid(gbm.maxBins,[20,30])
          .addGrid(gbm.maxIter,[10,20])
          .build())
cv=CrossValidator(estimator=gbm,estimatorParamMaps=paramGrid,evaluator=evaluator,numFolds=10)
```

```python
cv_model=cv.fit(train_df)
```

```python
y_pred=cv_model.transform(test_df)
ac=y_pred.select("label","prediction")
```

```python
ac.filter(ac.label==ac.prediction).count()/ac.count()# accuracy degerine ulasmıs oldık
```

```python
# auc degerine erismek istersek;(egri altında kalan alandır)
#(sınıflandırma da model basarısını degerlendirmek için kullanılır)
evaluator.evaluate(y_pred)
```
